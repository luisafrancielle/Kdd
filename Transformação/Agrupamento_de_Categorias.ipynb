```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando o arquivo xlsx
nome_do_arquivo = '12.xlsx'
df = pd.read_excel(nome_do_arquivo)
```

```python
#Substituindo valores de DS_TIPO_PREST_SOLICITANTE
valores_substituir = ['home care', 'cooperativa', 'intercambio', 'clinica de radiologia']
df['DS_TIPO_PREST_SOLICITANTE'].replace(valores_substituir, 'outros', inplace=True)
```

```python
#Substituindo valores de DS_INDICACAO_ACIDENTE
valores_substituir = ['transito', 'trabalho', 'outros']
df['DS_INDICACAO_ACIDENTE'].replace(valores_substituir, 'acidente', inplace=True)
```

```python
#Plotando os gráficos com os valores atualizados
contagem_categorias_dps = df['DS_TIPO_PREST_SOLICITANTE'].value_counts()
contagem_categorias_dps.plot(kind='barh', color='blue')

plt.xlabel('Categorias')
plt.ylabel('Frequência')
plt.title('Frequência de Categorias em DS_TIPO_PREST_SOLICITANTE com alterações')

plt.show()

contagem_categorias_dps = df['DS_INDICACAO_ACIDENTE'].value_counts()
contagem_categorias_dps.plot(kind='barh', color='blue')

plt.xlabel('Categorias')
plt.ylabel('Frequência')
plt.title('Frequência de Categorias em DS_INDICACAO_ACIDENTE com alterações')

plt.show()
```

```python
#Exportando o arquivo
df.to_excel('13.xlsx', index=False)
```
